```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = '/content/table of comparing.xlsx - Sheet4(5).csv'
df = pd.read_csv(file_path)
```

boxplot

```python
def plt_boxplot(measure_type):

    filtered_data = df[(df['Measure_Type'] == measure_type)]
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Measure', y='Value', hue="Dataset", data=filtered_data, width=.3,
                    notch=True, showcaps=False,
                    flierprops={"marker": "x"},
                    medianprops={"color": "r", "linewidth": 2})

    plt.title(f'Boxplot for {measure_type} Values for Each Dataset')
    plt.xlabel('Metrics')
    plt.ylabel('Values')
    plt.show()
```

Spectral

```python
measure_type = 'Spectral'
plt_boxplot(measure_type)
```

Semantic

```python
measure_type = 'Semantic'
plt_boxplot(measure_type)
```

**t-test**

```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel



def t_stat(df, Dataset_1, Dataset_2, Measure_Type, Measure):

    # Filter data
    dataset_1_values = df[(df['Dataset'] == Dataset_1) & (df['Measure_Type'] == Measure_Type) & (df['Measure'] == Measure)]['Value']
    dataset_2_values = df[(df['Dataset'] == Dataset_2) & (df['Measure_Type'] == Measure_Type) & (df['Measure'] == Measure)]['Value']

    # Perform one-sample paired t-test
    #t_stat, p_value = ttest_rel(ESC_50_values, another_dataset_values)
    t_stat, p_value = ttest_ind(dataset_1_values, dataset_2_values)

    return t_stat, p_value
```

```python
print('for Spectral and TV:')

print('\n')
result = t_stat(df, 'ESC-50', 'AudioGen', 'Spectral', 'TV')
print("Compare ESC-50 and AudioGen:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM_GPT', 'Spectral', 'TV')
print("Compare ESC-50 and AudioLDM_GPT:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM1', 'Spectral', 'TV')
print("Compare ESC-50 and AudioLDM1:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM2', 'Spectral', 'TV')
print("Compare ESC-50 and AudioLDM2:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")
```

```python
print('for Spectral and avg_pdist:')

print('\n')
result = t_stat(df, 'ESC-50', 'AudioGen', 'Spectral', 'avg_pdist')
print("Compare ESC-50 and AudioGen:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM_GPT', 'Spectral', 'avg_pdist')
print("Compare ESC-50 and AudioLDM_GPT:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM1', 'Spectral', 'avg_pdist')
print("Compare ESC-50 and AudioLDM1:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")

print('\n')
result = t_stat(df, 'ESC-50','AudioLDM2', 'Spectral', 'avg_pdist')
print("Compare ESC-50 and AudioLDM2:" f" t-statistic: {result[0]}" f", p-value: {result[1]}")
```
